```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Bowler's overall season's analysis

```python
df = pd.read_csv('/content/BOWLERS.csv')
df.head()
```

```python
average_runs_conceded = df['Runs'].mean()
average_wickets_taken = df['Wkts'].mean()
print("Average Runs Conceded:", average_runs_conceded)
print("Average Wickets Taken:", average_wickets_taken)
```

```python
# Find the bowler with the best economy
best_economy_bowler = df[df['Econ'] == df['Econ'].min()]
print("Bowler with Best Economy:")
best_economy_bowler
```

```python
# Find the bowler with the most wickets
most_wickets_bowler = df[df['Wkts'] == df['Wkts'].max()]
print("Bowler with Most Wickets:")
most_wickets_bowler
```

```python
# Sort DataFrame by WICKETS in descending order to get top WICKET TAKERS
df_top_WT = df.sort_values(by='Wkts', ascending=False).head(10)

# Horizontal bar chart for Top Run Scorers
plt.barh(df_top_WT['Player'], df_top_WT['Wkts'])
plt.title('Top 10 wicket takers')
plt.xlabel('wickets')
plt.ylabel('bowlers')
plt.show()
```

```python
# Plotting a scatter plot of economy vs wickets
plt.scatter(df['Econ'], df['Wkts'], marker='o')
plt.title('Economy vs Wickets')
plt.xlabel('Economy')
plt.ylabel('Wickets')
plt.show()
```

3. batsman's performance analysis (in 2021)

```python
df = pd.read_csv('/content/Most Runs - 2021.csv')
df.head()
```

```python
average_runs = df['Runs'].mean()
print("Average Runs :", average_runs)
average_SR = df['SR'].mean()
print("Average SR :", average_SR)
```

```python
# Find the BATSMAN with the highest SR
Highest_SR = df[df['SR'] == df['SR'].max()]
print("Batsman with Highest SR:")
Highest_SR
```

```python
# Find the BATSMAN with the highest runs in IPL2021
Highest_runs = df[df['Runs'] == df['Runs'].max()]
print("Batsman with Highest Runs:")
Highest_runs
```

```python
# Find the BATSMAN having the MAXIMUM SIXES
sixes = df[df['6s'] == df['6s'].max()]
print("Batsman with max sixes:")
sixes
```

```python

# Find the BATSMAN having the MAXIMUM fours
fours = df[df['4s'] == df['4s'].max()]
print("Batsman with max fours:")
fours
```
